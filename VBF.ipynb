# Medula

### Получаем названия нейронов из файла 

```python
def get_name_neuron(file_name):
    lst_FBbt = []
    file1 = open(file_name, "r")
    for i in range(250):
        line = file1.readline()
        if 'FBbt' in line:
            lst_FBbt.append(line.split('"')[1])
    file1.close
    return lst_FBbt
```

```python
lst_ME = get_name_neuron("237_Neurons_with_some_part_in_medulla.csv")
lst_ME
```

### Получаем symbols нейронов

```python
summ_ME = vc.neo_query_wrapper.get_type_TermInfo(lst_ME, summary = True)
```

```python
def get_symbol(FBbt):
    return (vc.neo_query_wrapper.get_type_TermInfo([FBbt], summary = True))[0]['symbol']
```

```python
lst_ME_symbols = [get_symbol(lst_ME[i]) for i in range(len(lst_ME))]
```

```python
np.unique(np.array(lst_ME_symbols))
```

### Строим граф групп нейронов для medulla

```python
def get_parents(FBbt, lst_region):
    res=[]
    parents = (vc.neo_query_wrapper.get_type_TermInfo([FBbt],summary=True)[0]['parents_id']).split('|')
    for parent in parents:
        if parent in lst_region:
            res.append(parent)
    return res
```

```python
get_parents('FBbt_00003826', lst_ME)
```

```python
df = pd.DataFrame(columns=['child', 'parent'])
df
```

```python
for FBbt in lst_ME:
    for parent in get_parents(FBbt, lst_ME):
        df = df.append({'child':FBbt, 'parent':parent}, ignore_index=True)
```

### Строим граф из relationships для Medulla

```python
def get_relationships(FBbt, lst_region):
    res=[]
    relations = (vc.neo_query_wrapper.get_type_TermInfo([FBbt])[0]['relationships'])
    for relation in relations:
        obj = relation['object']['short_form']
        if obj in lst_region:
            res.append(obj)
    return res
```

### Строим граф из relationships для Medulla и других нейронов

```python
def get_relationships_all(FBbt):
    res=[]
    relations = (vc.neo_query_wrapper.get_type_TermInfo([FBbt])[0]['relationships'])
    for relation in relations:
        obj = relation['object']['short_form']
        res.append(obj)
    return res
```

```python
df = pd.DataFrame(columns=['n1', 'n2'])
df
```

```python
def append_all_pairs(lst2D, FBbt):
    
    for relation in get_relationships_all(FBbt):
        #Если такой нейрон уже обрабатывался
        for pair in lst2D:
            if pair[0]==relation:
                return
        #Если такой нейрон еще не обрабатывался, то добавляем его и смотрим на его соседа
        lst2D.append([FBbt, relation])
        append_all_pairs(lst2D, relation)
```

```python
lst2D = []
for FBbt in lst_ME:
    append_all_pairs(lst2D,FBbt)
```
